## Working Locally

From an open terminal, you'll first get the class repo using `git`:
```bash
git clone https://github.com/profjsb/python-seminar.git

cd python-seminar

# create a conda environment called ay250_env
conda env create --prefix ~/ay250_env -f environment.yml
```

Activate the conda environment:
```bash
conda activate ~/ay250_env

```

Now fire up the Jupyter notebook
```bash
cd DataFiles_and_Notebooks
jupyter notebook
```

## Updating the environment

Later on, if you want to **update** the conda environment you can `pip install` or `conda install`. You can also edit the `environment.yml` file and then:
```bash
conda env update --prefix ~/ay250_env -f environment.yml --prune
```
To get out of the `conda` env:
```
deactivate
```
