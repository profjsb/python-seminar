$\textbf{Example 1: Super Bowl results.}$  ======================================================I have grabbed the results of the first 45 Super Bowls from http://www.pro-football-reference.com/Question of interest: based on the winning team's final score in a Super Bowl, can we reliably predict the margin of victory?Let's compute some summary statistics, fit a linear regression model, and check the goodness of that model fit.

```python
%pylab inlineimport numpyimport scipy.stats as stats# load results from all superbowlssb = numpy.loadtxt("superbowl_results.txt", delimiter=",", skiprows=1 \                   ,usecols=(3,5),dtype={'names': ('winnerpts','loserpts')\                                         ,'formats': ('i4','i4')})margin = sb['winnerpts']-sb['loserpts'] # margin of victoryprint('Scores of the last 5 Super Bowls (before yesterday):')print(sb[0:5])
```

```python
# plot winner points versus margin of victoryplt.figure(figsize=(10, 8))plt.plot(sb['winnerpts'],margin,'o')plt.xlabel('Winner Points')plt.ylabel('Margin of Victory')plt.title('Results of first 45 Super Bowls')
```

Let's first compute the correlation coefficient.  In scipy.stats we have several options, including Pearson's $r$, Spearman's $\rho$, and Kendall's $\tau$.

```python
# Pearson's rstats.pearsonr(sb['winnerpts'],margin)
```

```python
# Spearman's rhostats.spearmanr(sb['winnerpts'],margin)
```

```python
# Kendall's taustats.kendalltau(sb['winnerpts'],margin)
```

```python
# linear regressionslope, intercept, r_value, p_value, std_err = stats.linregress(x=sb['winnerpts'],y=margin)# calculate residualspred_lr = intercept + slope*sb['winnerpts']resids = margin - pred_lr# MSEn = len(resids)sigma_hat = sqrt(1./(n-2) * sum(resids**2))print("sigma-hat = " + str(sigma_hat))# plot regression lineplt.figure(figsize=(10, 8))plt.plot(sb['winnerpts'],margin,'o')plt.xlabel('Winner Points')plt.ylabel('Margin of Victory')plt.title('Results of first 45 Super Bowls')plt.plot((min(sb['winnerpts']),max(sb['winnerpts'])), \         (intercept + slope*min(sb['winnerpts']), intercept + slope*max(sb['winnerpts'])),\         'r-',linewidth=3)
```

```python
# histogram of residualsplt.figure(figsize=(10, 8))hist(resids,bins=7,color='b',normed=True)xlabel('Residuals')ylabel('Density')title('Histogram of Margin of Victory Residuals')xvec = arange(-15,15,.1)plot(xvec,map(lambda x: stats.norm.pdf(x,0,std(resids)),xvec),'r-',linewidth=3)legend(('N(0,$\sigma$)','Residuals'))
```

```python
# probability plot (aka QQ-plot)plt.figure(figsize=(10, 8))res_qq = stats.probplot(resids,(0.,std(resids)),'norm',plot=plt)
```

```python
# KS test for normalityks_pval = stats.kstest(resids,'norm',(0.,std(resids)))[0]print("KS-test for normality: p-value = " + str(ks_pval))
```

Finally, we will use this model to predict the Giants' margin of victory in Super Bowl XLVI given that they scored 21 Points.

```python
# linear regression predicted margin of victorywinscore_xlvi = 21pred_margin_xlvi = intercept + slope*winscore_xlviprint("Predicted Margin of Victory: " + str(round(pred_margin_xlvi)))
```

```python
# 90% prediction intervalSE_mu = sigma_hat * sqrt(1./n + (winscore_xlvi - sb['winnerpts'].mean())**2 \                         / sum((sb['winnerpts'] - sb['winnerpts'].mean())**2))alpha = 0.1SE_pred = stats.t.ppf(1-alpha/2,n-2)* sqrt( sigma_hat**2 + SE_mu**2)print('95% Prediction Interval: (' + str(pred_margin_xlvi - SE_pred) + ',' + \      str(pred_margin_xlvi + SE_pred) + ')')
```

$\textbf{Example 2: Super Bowl plays.}$  ======================================================I grabbed the data for every play in the Super Bowl for the last 4 years, including the number of yards gained, whether it was a run or a pass, the down, and the quarter of play.

```python
sbplays = numpy.loadtxt("superbowl_plays.txt", delimiter=",", \                        skiprows=1,usecols=(8,9,10,12,14),\                        dtype={'names': ('quarter','down','disttogo','yards','passer')\                               ,'formats': ('i4','i4','i4','i4','S16')})N = len(sbplays['yards'])
```

```python
## plot boxplotsplt.figure(figsize=(8, 6))boxplot(sbplays['yards'])ylabel('Yards Gained')
```

```python
# parallel boxplotsplt.figure(figsize=(10, 8))datalist = [sbplays['yards'][sbplays['quarter']==1],sbplays['yards'][sbplays['quarter']==2],\            sbplays['yards'][sbplays['quarter']==3], sbplays['yards'][sbplays['quarter']==4]]boxplot(datalist)xlabel('Quarter')ylabel('Yards Gained')
```

```python
### Kernel Density Estimationrunyards = np.array(sbplays['yards'][sbplays['passer']==''], dtype=np.float)passyards = np.array(sbplays['yards'][sbplays['passer']!=''], dtype=np.float)de_run = stats.gaussian_kde(runyards)de_pass = stats.gaussian_kde(passyards)yardvec = arange(-20.,80.,0.5)plt.figure(figsize=(10, 8))plot(yardvec,de_run.evaluate(yardvec),'r-',linewidth=3)plot(yardvec,de_pass.evaluate(yardvec),'b-',linewidth=3)legend(('Run','Pass'))xlabel('Yards')ylabel('Density')title('KDE for Super Bowl Plays')
```

```python
plt.figure(figsize=(10, 8))plot(yardvec,de_run.evaluate(yardvec),'r-',linewidth=3)plot(yardvec,de_pass.evaluate(yardvec),'b-',linewidth=3)legend(('Run','Pass'))xlabel('Yards')ylabel('Density')title('KDE for Super Bowl Plays')# KDE classifier run versus passfill_between(yardvec,0,0.0999,where=de_run.evaluate(yardvec) < de_pass.evaluate(yardvec)\             ,color='b',alpha=0.4)fill_between(yardvec,0,0.0999,where=de_run.evaluate(yardvec) >= de_pass.evaluate(yardvec)\             ,color='r',alpha=0.4)classify_pass = yardvec[de_run.evaluate(yardvec) < de_pass.evaluate(yardvec)]classify_run = yardvec[de_run.evaluate(yardvec) > de_pass.evaluate(yardvec)]print('Range Classified as Run: [' + str(min(classify_run)) + ',' \      + str(max(classify_run)) + '] yards')
```
